```python
# Ejercicio 1

# 1. Quiero un programa que guarde productos en un inventario
# Pasos a seguir: 
    # 1. Normalizar el nombre, minúsculas sin espacios
    # 2. Convertir la cantidad en un número
    # 3. Buscar si el producto existe en el inventario
    # 4. Si existe:sumar cantidad, actualizar precio 
    # 5. Si no existe: crear el producto, añadirlo al inventario


inventory= []
def add_product (name, price, quanty):
    name= name.strip().lower() # uso esto para evitar que se reconozca como el mismo producto a pesar de mayus/minus
    quanty= int(quanty) # convierto cantidad a número
    #busco si el producto existe
    for product in inventory: #buscar en el inventario a ver si el producto existe
        if product ["name"]== name: #si existe actualiza la cantidad pero no crea uno nuevo
        #si el producto existe, actualizar cantidad
            product ["quanty"]+= quanty #suma la cantidad nueva a la anterior
            product["price"] = price #actualiza al nuevo precio, siempre el últio será el que va
            return
    # si el producto no existe lo crea
    inventory.append ({
        "name":name,
        "price": price,
        "quanty": quanty        
    })
    print(f"Product {name} added: {quanty} units at ${price} each.")
    
# Ejercicio 2 - VER EL INVENTARIO

  # Quiero mostrar todos los productos que hay en el inventario
  # Debería recorrer todos los productos del invetario y mostrar: nombre, precio y cantidad
  # Mostrarlo de forma clara para que el cliente pueda ver toda la información del producto
  
def display_inventory():
    for product in inventory:
        print(f"Name: {product['name']}, Price: ${product['price']}, Quantity: {product['quanty']}")

# Ejercicio 3 - BUSCAR PRODUCTO
  # Quiero que la función me permita al cliente buscar un producto
  # Recorrer el inventario
  # Comparar cada producto con el nombre proporcionado por el cliente
  # Si lo encuentro avisrle al cliente y terminar
  # Si no lo encuentro informar al cliente que no existe

def find_product(name):
    found = False

    for product in inventory: # recorro en el inventario
        product_name = product["name"].strip().lower() #con strip elimino los espacios, y con lower convierto a minusculas para que eso no genere error
        search_name = name.strip().lower() # comparo el nombre en el inventario con el que busca el cliente

        if product_name == search_name:
            print("Product '" + name + "' is in the inventory")
            found = True #cuando lo encuentra pasa a ser true
            break

    if found == False:
        print("Product '" + name + "' is not in the inventory")


# Ejercicio 5 - ELIMINAR UN PRODUCTO

# Quiero eliminar un producto del sistema
# Informar al cliente que se eliminó
# Recorrr el inventario y comparar cada producto con el nombre igresado
# Si coincide: eliminar al producto e informar al cliente
# Terminar
# Si no se encuentra: avisar al cliente 

def delete_product (name):
    for product in inventory:
        if product ["name"]== name:
            inventory.remove (product)
            print (f"Product'{name}' delete ")


# Ejercicio 6 - CALCULAR EL VALOR DE UN PRODUCTO

# Calcular el valor total del inventario, para eso multiplico la cantidad que tengo de cada producto por su precio y sumo todo
# Mostrar el resultado al cliente
# Mi total inicia en 0
# Recorro cada producto y recojo el precio y la cantidad, multiplico y sumo al total
# Imprimir el total


def inventory_value ():
    total= 0
    for product in inventory:
        total+= product ["price"] * product ["quanty"]
    print (f"The total value is: ${total}")

```

```python
# Ejercicio 7 CARRITO DE LA COMPRA

# El comprador va a realizar varias acciones, necesitaré un menú que se repita y que se detenga cuando el cliente salga.

# Menú opciones para que el cliente elija una, me pregunto primero como elige si no sabe que productos hay entonces puedo usar la función de mostrar el inventario que hice antes.
# 1 Muestro los productos disponibles
# 2 Le pido al cliente el nombre del producto que quiere comprar
# 3 Le pregunto al cliente la cantidad
# 4 Debemos comprobar si lo que pide el cliente existe, para eso recorro los productos del inventario y su cantidad y comparo la que quiere con la que hay.
    # Si hay: añado el producto al carro, resto la cantidad del inventario, aviso que se añadió.
    # Si no hay suficiente:  aviso cuantas unidades hay disponibles
    # Si no hay hay que avisarle al cliente, por ejemplo producto no encontrado.
# 5 Dar opción al cliente de que vea su carrito para ver lo que está comprando:
    # Avisar si el carrito está vacío
    # Recorro cada producto del carrito
    # Calcular subtotal, precio, cantidad, ir sumando y mostrar todo al cliente
# 6 Se termina cuando el cliente elige salir para eso muestro un mensaje de despedida y se rompe el ciclo
```

```python
# Defino la función que contiene el funcionamiento del carrito
def shopping_cart():
    cart = [] #lista para almacenar los productos que el cliente va añadiendo al carrito

    while True: # Bucle infinito para mostrar el menú hasta que el cliente decida salir
        # Muestra el menú de opciones al cliente
        print("\n- Shopping Cart-") # \n para indicar el salto de línea
        print(" 1. Add product to cart")
        print(" 2. View cart")
        print(" 3. Exit")
        # Pide al cliente que elija una opción, el strip es para eliminar espacios al principio y al final por si el cliente se equivoca y mete espacios
        choice = input(" Choose an option: ").strip()

        if choice == "1": #Si el cliente elige la opción 1, añadirá un producto al carrito
            print("\nAvailable products:") #Muestra los productos disponibles usando: display_inventory() creda anteriormente
            display_inventory()
            name = input("Product name: ").strip()
            quanty = int(input("Quanty: "))

            found = False # Variable para verificar si el producto se encuentra en el inventario (lo marco como false porque aún nosé si está)
            for product in inventory: # Recorro el inventario para buscar el producto que el cliente quiere comprar
                if product["name"].lower() == name.lower():
                    found = True

                    if quanty <= product["quanty"]: # Verifico si hay suficiente cantidad del producto en el inventario
                        # añadir al carrito
                        cart.append({
                            "name": product["name"], # Podría habrerlo hecho con clases tb preguntar Yanelis
                            "price": product["price"],
                            "quanty": quanty
                        })

                        # restar producto añadido del inventario, tomo la cantidad actual y le resto lo que el cliene pone y guardo nuevo valor
                        product["quanty"] -= quanty 

                        print(f"{quanty} {product["name"]} added to cart")
                    else:
                        print(f"Not enough stock. Available: {product["quanty"]}") # F me permite insetar variable dentro del string

                    break

            if not found:
                print(f"Product '{name}' not found")

        elif choice == "2": # Aquí el cliente puede ver su carrito, si elige la opción 2
            if not cart:
                print("Your cart is empty")
            else:
                total = 0
                print("\n - Your Cart-")

                for item in cart: # Recorro cada producto del carrito para mostrarlo al cliente, calcular el subtotal y el total
                    subtotal = item["price"] * item["quanty"]
                    total += subtotal
                    print(f"{item["name"]} - ${item["price"]} x {item["quanty"]} = ${subtotal}")

                print(f"Total: $ {total}")

        elif choice == "3":
            print("Hasta la vista, baby")
            break

        else:
            print("Try again")
```
